```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# Load your CSV file into a pandas DataFrame
df = pd.read_csv('/content/drive/MyDrive/FIRE_2023/AISoME/train.csv')

# Split the labels into a list of labels
df['labels'] = df['labels'].str.split(' ')

# Create an instance of MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Fit the MultiLabelBinarizer on your labels column
mlb.fit(df['labels'])

# Transform the labels column into a binary matrix
label_matrix = mlb.transform(df['labels'])

# Create a DataFrame from the binary matrix
label_df = pd.DataFrame(label_matrix, columns=mlb.classes_)

# If 'none' label is present in the columns, set all its values to 0
if 'none' in label_df.columns:
    label_df['none'] = 0

# Concatenate the original DataFrame with the label DataFrame
result_df = pd.concat([df, label_df], axis=1)
result_df = result_df.drop(['ID', 'none'], axis=1)

column_Order =  ('tweet', 'side-effect', 'ineffective', 'rushed', 'pharma', 'mandatory', 'unnecessary', 'political', 'ingredients', 'conspiracy', 'country', 'religious')

result_df = result_df.reindex(columns=column_Order)
result_df.head(10)
# Save the resulting DataFrame to a new CSV file
result_df.to_csv('/content/drive/MyDrive/FIRE_2023/AISoME/binarized_raw_dataset.csv', index=False)
```
